```python
def domneva(G):
    if G.is_connected() == False:
        return None
    B = set(G.periphery())
    d = G.distance_all_pairs()
    razdalja = []
    for v in G:
        if v not in B:
            sezv = []
            for u in B:
                sezv.append(d[v][u])
            razdalja.append(min(sezv))
    razlika = 2 * G.dominating_set(total=True, value_only=True) - G.radius() - (max(razdalja) if razdalja else 0)
    if razlika < 0:
        return False
    else: return razlika
```
